```python
#This script plots the leakage current of the detector strips and guard ring on the same graph.
## Note this script only works on Windows OS (specifically the path part) ###
### but import os should allow this code to be modified to work on any OS ###

import matplotlib.pyplot as plt
import pylab


################ CHANGE THESE PARAMETERS WITH DETECTOR ################################
temp = "-42C" # "-42C"
if temp == "RT":
    temp_factor = 1
    units = "Microamperes"
elif temp == "-42C":
    temp_factor = 1000
    units = "Nanoamperes"
print("Units: "+units)

#Path to the file below (obviously change for different machines)
prename = r"C:\Users\willi\Dropbox (MIT)\Perez Lab GAPS Dropbox\Si(Li) Testing\Data\Sh0149\IV Curves\2025\03-08-2025" 
pfx = "\\Sun_"                                  #Use the prefix for the file here, like PS or Sun #pfx = "\\PS_"
#pfx = "\\PS_"                                  #Probe station prefix
postname = "_-42C"
detno = "Sh0149"

title = detno + " All Strips IV Curve " + postname       #Name the plot title displayed on the plot
filename = "Allstrips" + postname               #Give the png file a name, the file will be named detno + " " + filename
savename = prename + "\\" + filename            #Direct the save location

#lets = ['BTop','CTop','CBot','DTop','DBot','ETop','EBot','F1Top','GBot','RBot','RTop','H1Bot','H2Bot','A1Top','A2Top','F1Bot','F2Bot']
lets = ['A','B','C','D','E','F','G','H','R']    #8-strip detector
#lets = ['A','B','C','D','E','F','G','R']        #7-strip detector
#lets = ['A','R']                                #1-strip detector
#lets = ['A','B','C','D','R']                   #4-strip detector

#cols = ['red',red','red','red',red','red','red',red','red'] 
#cols = ['green','tan', 'tan','cyan','cyan','navy','navy','violet','teal','goldenrod','darkgoldenrod','gray','gray','blue','darkblue','saddlebrown','sienna'] #If you would like colors for any reason you can choose them here

####################### END CHANGE SECTION ############################################

plt.figure(figsize=(6, 4),dpi=100)
plt.yscale('log')
plt.xlabel("HV Bias (V)")
plt.ylabel("Leakage Current (uA)")
#plt.ylim(1e-5,1000)
plt.title(title)

for j, x in enumerate(lets):
    #print("hi")
    name = prename + pfx + lets[j] + postname
    nlines = len(open(name).readlines(  ))
    file = open(name, "r")
    tarray = [0]*3
    V = [0]*nlines #Could also probably accomplish this within the array with a function that adds another element to the array.
    I = [0]*nlines
    
    for i, x in enumerate(file):
        #print(x.strip().split(','))
        tarray = x.strip().split(',')
        #print(i)
        V[i] = float(tarray[1])
        I[i] = float(tarray[2])
        
    print(lets[j] + ": " + str(round(I[10]*temp_factor, 2)))
    plt.plot(V, I, 'o', label = lets[j]) 
    #plt.plot(V, I, cols[j], label = lets[j]) #Can uncomment this if for whatever reason you'd like colors on your plot. 
    file.close()

plt.legend(loc='upper left')
plt.savefig(savename) #Save the file
plt.show() #Display the plot
```

```python
# FOR WINDOWS!!!!
# Delta plots DATE2 - DATE1

# This script plots the leakage current of the detector strips and guard ring on the same graph.
# import os should allow this code to be modified to work on any OS

import matplotlib.pyplot as plt
import pylab

################ CHANGE THESE PARAMETERS WITH DETECTOR ################################
temp = "-42" # "-42"
if temp == "RT":
    temp_factor = 1
    units = "Microamperes"
elif temp == "-42":
    temp_factor = 1000
    units = "Nanoamperes"
print("Units: "+units)

detno = "Sh0149"
date1 = "03-07-2025" #03-09-2023, 02-20-2024
year1 = "2025" #2023, 2024
date2 = "03-08-2025" 
year2 = "2025"
detno = "Sh0149"

#"C:\Users\willi\Dropbox (MIT)\Perez Lab GAPS Dropbox\Si(Li) Testing-checkpoint.ipynb"
#Path to the file below (obviously change for different machines)
prename = r"C:\Users\willi\Dropbox (MIT)\Perez Lab GAPS Dropbox\Si(Li) Testing\Data"
prename += "\\" + detno + r"\IV Curves"
prename1 = prename+ "\\" + year1 + "\\" + date1
prename2 = prename+ "\\" + year2 + "\\" + date2

pfx = r"\Sun_"                                  #Use the prefix for the file here, like PS or Sun #pfx = "\\PS_"
pfx2 = r"\Sun_"                                  #Probe station prefix
postname1 = "_-42C"
postname2 = "_-42C"

title = date2 + " - " + date1 + " Delta Plot" + detno + " All Strips IV Curve " + postname1       #Name the plot title displayed on the plot
#filename = "Allstrips" + postname               #Give the png file a name, the file will be named detno + " " + filename
#savename = prename + "\\" + filename            #Direct the save location

lets = ['A','B','C','D','E','F','G','H','R']    #8-strip detector
#lets = ['A','R']                               #1-strip detector
#lets = ['A','B','C','D','R']                  #4-strip detector

#cols = ['ro', 'ro','ro','ro','ro','ro','ro','ro','ro'] #If you would like colors for any reason you can choose them here

####################### END CHANGE SECTION ############################################

plt.figure(figsize=(6, 4),dpi=100)
#plt.yscale('symlog')
plt.xlabel("HV Bias (V)")
plt.ylabel("Leakage Current (uA)")
#plt.ylim(-10,1000)
plt.title(title)

for j, x in enumerate(lets):
    name1 = prename1 + pfx + lets[j] + postname1
    name2 = prename2 + pfx2 + lets[j] + postname2
    nlines1 = len(open(name1).readlines(  ))
    nlines2 = len(open(name2).readlines(  ))
    file1 = open(name1, "r")
    file2 = open(name2, "r")
    tarray = [0]*3
    V = [0]*nlines1 #Could also probably accomplish this within the array with a function that adds another element to the array.
    I1 = [0]*nlines1
    I2 = [0]*nlines2
    
    for i, x in enumerate(file1):
        #print(x.strip().split(','))
        tarray = x.strip().split(',')
        #print(i)
        V[i] = float(tarray[1])
        I1[i] = float(tarray[2])
    for i, x in enumerate(file2):
        #print(x.strip().split(','))
        tarray = x.strip().split(',')
        #print(i)
        I2[i] = float(tarray[2])
    #print(type(I1))
    #print(type([a_i - b_i for a_i, b_i in zip(I1, I2)]))
    #print(diffI)
    print(lets[j] + ": " + str(round(I1[10]*temp_factor, 2)) + " --> " +str(round(I2[10]*temp_factor, 2)))
    diffI = [a_i - b_i for a_i, b_i in zip(I2, I1)]
    plt.plot(V, diffI, 'o', label = lets[j]) 
    #plt.plot(V, I, cols[j], label = lets[j]) #Can uncomment this if for whatever reason you'd like colors on your plot. 
    file1.close()
    file2.close()
    
plt.legend(loc='upper left')
# print(V1)
# print(I1)
# print(V2)
# print(I2)
plt.legend(loc='upper left')
#plt.savefig(savename) #Save the file
plt.show() #Display the plot
```
